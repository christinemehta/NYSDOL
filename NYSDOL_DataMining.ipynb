# Creating a new Dataframe for further analysis

```python
Clean = Data[['NAICS','AREATYPE', 'AREA', 'YEAR', 'AVGEMP']]
```

# Creating the list of years to be analyzed

```python
years = []
for i in range(2000,2018):
    years.append(i)
```

```python
Areatype = ["County"]
Area = ['Tompkins County','Tompkins', 'Tioga County', 'Steuben County', 'Schuyler County','Delaware County','Chenango County','Chemung County','Broome County']
naics = ['112','111','113','114','115','311','312']
SouthernTier = Clean[Clean.YEAR.isin(years) & Clean.AREATYPE.isin(Areatype) & Clean.AREA.isin(Area) & Clean.NAICS.isin(naics)]
SouthernTier
```

# Southern Tier Region

```python
"""
This is for the Southern Tier Region in Upstate NY over the period of 2000 - 2017.
"""
YEars = []
sum0 = []
for i in years:
    YEars.append(i)
    Areatype = ["County"]
    Area = ['Tompkins County','Tompkins', 'Tioga County', 'Steuben County', 'Schuyler County','Delaware County','Chenango County','Chemung County','Broome County']
    naics = ['112','111','113','114','115','311','312']
    SouthernTier = Clean[Clean.YEAR.isin(YEars) & Clean.AREATYPE.isin(Areatype) & Clean.AREA.isin(Area) & Clean.NAICS.isin(naics)]
    print("For the year " + str(YEars[0]) + ", the avg employment in the Southern Tier is",np.sum(SouthernTier['AVGEMP']))
    sum0.append(np.sum(SouthernTier['AVGEMP']))
    YEars = []
fig = plt.plot(years,sum0)
plt.xlabel("Years")
plt.ylabel("Average Employment")
plt.xlim(2000,2017,1)
```
